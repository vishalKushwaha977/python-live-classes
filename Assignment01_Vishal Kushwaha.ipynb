```python
# Pattern 5.
#       *
#      ***
#     *****
#    *******
#   *********
for i in range(1 , 6):
    #print(i)
    print(" " * (5 - i)+ "*" *(2* i -1))
```

```python
# Pattern 6.
#     *
#    ***
#   *****
#  *******
# *********
#  *******
#   *****
#    ***
#     *
for i in range (1 , 6):
    print(" " * (5 - i)+ "*" *(2 * i -1))
for i in range(4 , 0 , -1):
    print(" "* (5 -i)+ "*" * (2 * i -1))
```

```python
# Write a program that takes a number from the user and prints whether it is Even or Odd.
try:
    num = int(input("Enter your number:-"))
    if not num % 2:
        print("Your number is even")
    else:
        print("Your number is odd")
except:
    print("you have enter invalid number please try again")
```
