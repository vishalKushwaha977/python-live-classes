```python
def binary_search(num , arr):
    l = 0
    r = len(arr) -1
    while l <= r:
        m =(l + r)//2
        if num < arr[m]:
            r = m - 1
        elif num > arr[m]:
            l = m + 1
        else:
            return True,f"Value found and index: {m}"
    return "value not found in the array"
```

```python
a = [1,2,4,5,6,10,18,45]
n = int(input("Enter number to find"))
binary_search(n ,a)
```
