```python
#1572. Matrix Diagonal Sum
from typing import List
class Solution:
    def diagonalSum(self, mat: List[List[int]]) -> int:
        n = len(mat)
        total = 0
        for i in range(n):
            total += mat[i][i]
            if i != n -1 - i:
                total += mat[i][n - 1 - i]
        return total
obj = Solution()
result = obj.diagonalSum([[1,2,3],[4,5,6],[7,8,9]])
print(result)
```

```python
#2643. Row With Maximum Ones
class Solution:
    def rowAndMaximumOnes(self, mat: List[List[int]]) -> List[int]:
        ans = [0] * 2
        idx = 0 
        mxcnt = 0
        for i in range(len(mat)):
            cnt = 0
            for ele in mat[i]:
                if ele == 1:
                    cnt += 1
            if cnt > mxcnt:
                idx = i
                mxcnt = cnt
        ans[0] = idx
        ans[1] = mxcnt
        return ans
 
obj = Solution()
result = obj.rowAndMaximumOnes([[0,1],[1,0]])
print(result)
```
