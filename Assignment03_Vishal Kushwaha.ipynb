```python
# 20 
class Solution:
    def isValid(self, s: str) -> bool:
        stack = []
        pairs = {')': '(', '}': '{', ']': '['}
        for ch in s:
            if ch in pairs:
                if not stack:
                    return False
                if stack.pop() != pairs[ch]:
                    return False
            else:
                stack.append(ch)
       
        return not stack
    
obj = Solution()
result = obj.isValid("{[()]}")
print(result)
```

```python
#2000. Reverse Prefix of Word
class Solution:
    def reversePrefix(self, word: str, ch: str) -> str:
        for i in range(len(word)):
            if word[i] == ch:
                return word[i::-1] + word[i+1:]
        return word
        
obj = Solution()
result = obj.reversePrefix("abcdefd","d")
print(result)
```

```python
# 682. 
from typing import List
class Solution:
    def calPoints(self, operations: List[str]) -> int:
        stack = [] 
        for n in operations:
            if n == "+":
                stack.append(stack[-1] + stack[-2])
            elif n == "D":
                stack.append(2 * stack[-1])
            elif n == "C":
                stack.pop()
            else:
                stack.append(int(n))
        return sum(stack)

obj = Solution()
result = obj.calPoints(["5","2","C","D","+"])
print(result)
```

```python
#1021
class Solution:
    def removeOuterParentheses(self, s: str) -> str:
        result = []
        balance = 0
        for ch in s:
            if ch =='(':
                if balance > 0:
                    result.append(ch)
                balance += 1
            else:
                balance -= 1
                if balance > 0 :
                    result.append(ch)
        return "".join(result)
                    
        
obj = Solution()
result = obj.removeOuterParentheses( "(()())(())")
print(result)
```
