```python
#2730. Find the Longest Semi-Repetitive Substring
class Solution:
    def longestSemiRepetitiveSubstring(self, s: str) -> int:
        left = 0
        pairs = 0
        max_len = 1
        for right in range(1 , len(s)):
            #check adjacent pair
            if s[right] == s[right -1]:
                pairs += 1   
            # shrink window if more then one pair
            while pairs > 1:
                if s[left] == s[left + 1]:
                    pairs -= 1
                left += 1
            max_len = max(max_len , right - left + 1)
        return max_len
                            
obj = Solution()
result = obj.longestSemiRepetitiveSubstring("52233")
print(result)
```
