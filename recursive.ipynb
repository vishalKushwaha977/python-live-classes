```python
#17. Letter Combinations of a Phone Number
from typing import List
class Solution:
    def letterCombinations(self, digits: str) -> List[str]:
        mapping = {
            "2":"abc",
            "3":"def",
            "4":"ghi",
            "5":"jkl",
            "6":"mno",
            "7":"pqrs",
            "8":"tuv",
            "9":"wxyz"
        }
        res = []
        def backtracking(i , curr_str):
            if i == len(digits):
                res.append(curr_str)
                return
            letters = mapping[digits[i]]
            
            for letter in letters:
                backtracking(i + 1, curr_str + letter)
        backtracking(0,"")
        return res
obj = Solution()

result = obj.letterCombinations("23")
print(result)
```

```python
#509.Fibonacci Number
class Solution:
    def fib(self, n: int) -> int:
        if n == 0 or n == 1:
            return n
        return self.fib(n - 1) + self.fib(n -2)
obj = Solution() 
result  = obj.fib(3)
print(result)
```

```python
# tail recursion
def solve(n):
    if n  == 0:
        return 
    print(n)
    solve(n-1)
solve(4)
```

```python
# head recursion
def solve(n):
    if n == 0:
        return
    solve(n -1)
    print(n)
solve(4)
```

```python
#Tree recursion
def solve(n):
    if n == 0:
        return
    solve(n - 1)
    print(n)
    solve(n- 1)
solve(4)
```

```python
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None
# Recursive DFS Function
def dfs(Node):
    if Node is None:
        return 
    print("Visiting",Node.val)
    dfs(Node.left)
    dfs(Node.right)

root = TreeNode(1)
root.right = TreeNode(3)
root.left = TreeNode(2)
root.right.right = TreeNode(5)
root.left.left = TreeNode(4)


# run 🏃‍➡️🏃‍➡️
dfs(root)
  
```
