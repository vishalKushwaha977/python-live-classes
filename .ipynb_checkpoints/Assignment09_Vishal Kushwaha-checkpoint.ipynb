```python
#207. Course Schedule
from typing import List

class Solution:
    def canFinish(self, numCourses: int, prerequisites: List[List[int]]) -> bool:
        graph = {}
        
        for i in range(numCourses): 
            graph[i] = []
        for a, b in prerequisites:
            print(a,b)
            graph[b].append(a)
        cycle = set()
        visit = set()
        def dfs(node):
            if node in cycle:
                return False
            if node in visit:
                return True
            cycle.add(node)
            for nei in graph[node]:
                if not dfs(nei):
                    return False
            cycle.remove(node)
            visit.add(node)
            return True
            
        for i in range(numCourses):
            if not dfs(i):
                return False
        return True
        
        
numCourses = 2
prerequisites = [[1, 0],[0, 1]]
obj = Solution()
result = obj.canFinish(numCourses,prerequisites)

print("Can finish all courses:", result)
            
```

```python
#1311. Get Watched Videos by Your Friends
class Solution:
    def watchedVideosByFriends(self, watchedVideos: List[List[str]], friends: List[List[int]], id: int, level: int) -> List[str]:
        visited = set([id])
        queue = deque([id])

        for _ in range(level):
            for _ in range(len(queue)):
                person = queue.popleft()
                for f in friends[person]:
                    if f not in visited:
                        visited.add(f)
                        queue.append(f)
        videos = []
        for person in queue:
            videos.extend(watchedVideos[person])
        count = Counter(videos)
        return sorted(count.keys(), key=lambda x: (count[x], x))
```
