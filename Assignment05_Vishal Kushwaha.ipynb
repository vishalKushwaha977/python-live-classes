```python
#21
from typing import Optional
class Node:
    def __init__(self,val = 0, next = None):
        self.val = val
        self.next = None
class LinkedList:
    def mergeTwoLists(self, list1: Optional[Node], list2: Optional[Node]):
        dummy = Node(0)
        tail = dummy
        while list1 and list2:
            if list1.val <= list2.val:
                tail.next = list1
                list1 = list1.next
            else:
                tail.next = list2
                list2 = list2.next
            tail = tail.next
        tail.next = list1 if list1 else list2
        return dummy.next

a1 = Node(1)
a2 = Node(3)
a3 = Node(5)
a1.next = a2
a2.next = a3
list1 = a1

b1 = Node(2)
b2 = Node(4)
b3 = Node(6)
b1.next = b2
b2.next = b3
list2 = b1
        
sol = LinkedList()
merged_head = sol.mergeTwoLists(list1, list2) 
current = merged_head
while current:
    print(current.val, end=" -> ")
    current = current.next
print("None")
```

```python
# 160. Intersection of Two Linked Lists
class Node:
    def __init__(self,val = 0, next = None):
        self.val = val
        self.next = None
class LinkedList:
    def getIntersectionNode(self, headA: Node, headB: Node):
        lista = headA
        listb = headB
        while lista != listb:
            if lista is not None:
                lista = lista.next
            else:
                lista = headB
                
            if listb is not None:
                listb = listb.next
            else:
                 listb = headA
        return lista

intersect = Node(8)
intersect.next = Node(9)

headA = Node(1)
headA.next = Node(2)
headA.next.next = Node(3)
headA.next.next.next = intersect
headB = Node(4)
headB.next = Node(5)
headB.next.next = intersect
sol = LinkedList()
result = sol.getIntersectionNode(headA, headB)

if result:
    print("Intersection at node with value:", result.val)
else:
    print("No intersection")
```

```python
#203. Remove Linked List Elements
from typing import Optional
class Node:
    def __init__(self,val = 0, next = None):
        self.val = val
        self.next = None
class LinkedList:
    def removeElements(self, head: Optional[Node], val: int):
        dummy = Node(0)
        dummy.next = head
        
        current = dummy
        while current.next:
            if current.next.val == val:
                current.next = current.next.next
            else:
                current = current.next

        return dummy.next
            
head = Node(1)
head.next = Node(2)
head.next.next = Node(6)
head.next.next.next = Node(3)
head.next.next.next.next = Node(4)
head.next.next.next.next.next = Node(5)
head.next.next.next.next.next.next = Node(6)
ll = LinkedList()
new_head = ll.removeElements(head, 6)
current = new_head
while current:
    print(current.val, end=" -> ")
    current = current.next
print("None")
```

```python
#92. Reverse Linked List II
from typing import Optional

class Node:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


class LinkedList:
    def reverseBetween(self, head: Optional[Node], left: int, right: int) -> Optional[Node]:
        if not head or left == right:
            return head

        dummy = Node(0)
        dummy.next = head
        prev = dummy

        for _ in range(left - 1):
            prev = prev.next

        curr = prev.next
        for _ in range(right - left):
            temp = curr.next
            curr.next = temp.next
            temp.next = prev.next
            prev.next = temp

        return dummy.next
        
head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)

ll = LinkedList()
new_head = ll.reverseBetween(head, 2, 4)
curr = new_head
while curr:
    print(curr.val, end=" -> ")
    curr = curr.next
print("None")
```
