```python
#128. Longest Consecutive Sequence
from typing import List
class Solution:
    def longestConsecutive(self, nums: List[int]) -> int:
        num_set = set(nums)
        longest = 0
        print(num_set)
        for num in num_set:
            
            if num -1 not in num_set:
                current  = num 
                length = 1

                while current + 1 in num_set:
                    current += 1
                    length += 1
                longest = max(longest,length)
        return longest
            
obj = Solution()
result = obj.longestConsecutive([100,4,200,1,3,2])
print(result)
```

```python
#15
from typing import List
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        nums.sort()
        result = []
        n = len(nums)
    
        for i in range(n - 2):
            
            # skip duplicate first numbers
            if i > 0 and nums[i] == nums[i - 1]:
                continue
    
            left = i + 1
            right = n - 1
    
            while left < right:
                total = nums[i] + nums[left] + nums[right]
                if total == 0:
                    result.append([nums[i], nums[left], nums[right]])
    
                    # skip duplicates
                    while left < right and nums[left] == nums[left + 1]:
                        left += 1
                    while left < right and nums[right] == nums[right - 1]:
                        right -= 1
    
                    left += 1
                    right -= 1
    
                elif total < 0:
                    left += 1
                else:
                    right -= 1
    
        return result
obj = Solution()
result = obj.threeSum([-1,0,1,2,-1,-4])
print(result)
```

```python
#424
class Solution:
    def characterReplacement(self, s: str, k: int) -> int:
        count = {}          # frequency map
        left = 0
        maxFreq = 0         # max frequency in current window
        result = 0

        for right in range(len(s)):
            # add current character
            count[s[right]] = count.get(s[right], 0) + 1
            maxFreq = max(maxFreq, count[s[right]])

            # if more than k replacements needed, shrink window
            while (right - left + 1) - maxFreq > k:
                count[s[left]] -= 1
                left += 1

            # update result
            result = max(result, right - left + 1)

        return result
obj = Solution()
result = obj.characterReplacement("ABAB",2)
print(result)
```
