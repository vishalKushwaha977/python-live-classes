```python
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        data = {}
        
        for i, num in enumerate(nums):
            newNum = target - num
            #print(newNum)
            if newNum in data:
                print(i)
                return
                
            
            data[num] = i 
            print(data[num])

nums = [2, 7, 11, 15]
target = 9
obj = Solution()
result = obj.twoSum(nums, target)
#print(result)


```
