```python
#2073. Time Needed to Buy Tickets
from collections import deque
from typing import List
class Solution:
    def timeRequiredToBuy(self, tickets: List[int], k: int) -> int:
        queue = deque()
        ans = 0
        i = 0
        while i < len(tickets):
            queue.append([i , tickets[i]])
            i += 1
        while queue:
            i , v = queue.popleft()
            v -= 1
            ans += 1
            if v != 0:
                queue.append([i , v])
            if i == k and v == 0:
                return ans
        return ans

obj = Solution()
result = obj.timeRequiredToBuy([5,1,1,1],0)
print(result)  
```

```python
#225.Implement Stack using Queues
from collections import deque
class MyStack:
    def __init__(self):
        self.q1 = deque()
        self.q2 = deque()
        
    def push(self,x:int):
        self.q2.append(x)
        while self.q1:
            self.q2.append(self.q1.popleft())
        
        self.q1, self.q2 = self.q2 , self.q1
    
    def pop(self):
        return self.q1.popleft()
    def top(self):
        return self.q1[0]
    def empty(self):
        return len(self.q1) == 0
```

```python
obj = MyStack()
obj.push(10)
obj.top()
obj.pop()
obj.empty()
```

```python
#150. Evaluate Reverse Polish Notation
from typing import List
class Solution:
    def evalRPN(self, tokens: List[str]) -> int:
        stack = []
        for i in tokens:
            if i not in "+-*/":
                stack.append(int(i))
            else:
                b = int(stack.pop())
                a = int(stack.pop())
                if i == "+":
                    stack.append(a + b)
                elif i == "-":
                    stack.append(a - b)
                elif i == "*":
                    stack.append(a * b)
                else:
                    stack.append(int(a / b))
        return stack[0]
        
obj = Solution()
result = obj.evalRPN(["2","1","+","3","*"])
print(result)
```

```python
#71.Simplify Path

class Solution:
    def simplifyPath(self, path: str) -> str:
        stack = []
        for part in path.split('/'):
            if part == "" or part == ".":
                continue
            elif part == "..":
                if stack:
                    stack.pop()
            else:
                stack.append(part)
        return"/"+"/".join(stack)
obj = Solution()
result = obj.simplifyPath("/home/..//foo/")
print(result)  
```
