```python
#206. Reverse Linked List
class ListNode:
    def __init__(self,val =0 ,next = None):
        self.val = val
        self.next = next
class Solution:
    def print_List(self,head):
        curr = head
        while curr:
            print(curr.val,end="-->")
            curr = curr.next
        print("None")
       
    def reverseList(self,head):
        prev = None
        curr = head
        self.print_List(head)
        while curr:
            nxt = curr.next
            curr.next = prev
            prev = curr
            curr = nxt
        self.print_List(prev)
        
        return prev
```

```python
head = ListNode(1, ListNode(2, ListNode(3)))
sol = Solution()
sol.reverseList(head)
```
