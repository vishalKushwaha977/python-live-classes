```python
#11. Container With Most Water
# Formula Water = min(left_height, right_height) × distance_between_lines
class Solution:
    def maxArea(self,height:list[int]):
        left = 0
        right = len(height) - 1
        max_water = 0
        while left < right:
            width = right - left
            h = min(height[left] , height[right])
            area = width * h
            max_water = max(max_water, area)
            if height[left] < height[right]:
                left += 1
            else:
                right -= 1
        return max_water
        
            
obj = Solution()
result = obj.maxArea([1,8,6,2,5,4,8,3,7])
print(result)
```

```python
#20 Valid Parentheses
class Solution:
    def isValid(self, s: str) -> bool:
        stack = []
        pairs = {')': '(', '}': '{', ']': '['}

        for ch in s:
            if ch in pairs:                 # closing bracket
                if not stack:
                    return False
                if stack.pop() != pairs[ch]:
                    return False
            else:                            # opening bracket
                stack.append(ch)

        return not stack
    
obj = Solution()
result = obj.isValid("{[()]}")
print(result)
```

```python
#735 .Asteroid Collision
from typing import List
class Solution:
    def asteroidCollision(self, asteroids: List[int]) -> List[int]:
        stack = []

        for ast in asteroids:
            while stack and ast < 0 and stack[-1] > 0:
                diff = ast + stack[-1]

                if diff < 0:
                    stack.pop()      # stack asteroid destroyed
                elif diff > 0:
                    ast = 0          # current asteroid destroyed
                else:
                    ast = 0
                    stack.pop()      # both destroyed

            if ast:
                stack.append(ast)

        return stack


obj = Solution()
result = obj.asteroidCollision([5, 10, -5])
print(result)
```

```python
#150. Evaluate Reverse Polish Notation
from typing import List
class Solution:
    def evalRPN(self, tokens: List[str]) -> int:
        stack = []
        for n in tokens:
           if n not in "+-*/":
               stack.append(n)
           else:
               b = int(stack.pop())
               a = int(stack.pop())
               if n == "+":
                   stack.append(a + b)
               elif n == "-":
                   stack.append(a - b)
               elif n == "*":
                   stack.append(a * b)
               else:
                   stack.append(int(a / b))
        return stack[0]
                                   
obj = Solution()
result = obj.evalRPN(["2","1","+","3","*"])
print(result)
```

```python
#3775 Reverse Words With Same Vowel Count
#Determine how many vowels appear in the first word. Then, reverse each following word that has the same vowel count. 
#Leave all remaining words unchanged
class Solution:
    def reverseWords(self, s: str) -> str:
        vowels  = {"a","e","i","o","u"}
        first_cnt = 0 
        words = s.split()
    
        for ch in words[0]:
             if ch in vowels:
                 first_cnt += 1
        for i in range(1 , len(words)):
            count = 0 
            for c in words[i]:
                if c in vowels:
                    count += 1
            if count == first_cnt:
                words[i] = words[i][::-1]
        return " ".join(words)
                 
obj = Solution()
result = obj.reverseWords("cat and mice")
print(result)
```

```python
#2073. Time Needed to Buy Tickets
from collections import deque
from typing import List
class Solution:
    def timeRequiredToBuy(self, tickets: List[int], k: int) -> int:
        queue = deque()
        ans = 0
        i = 0
        while i < len(tickets):
            queue.append([i , tickets[i]])
            i += 1
        while queue:
            i , v = queue.popleft()
            v -= 1
            ans += 1
            if v != 0:
                queue.append([i , v])
            if i == k and v == 0:
                return ans
        return ans

obj = Solution()
result = obj.timeRequiredToBuy([5,1,1,1],0)
print(result)  
```

```python
#2535 Difference Between Element Sum and Digit Sum of an Array
from typing import List
class Solution:
    def differenceOfSum(self, nums: List[int]) -> int:
        ele_sum = sum(nums)
        digit_sum = 0
        for num in nums:
            temp = num
            while num > 0:
                temp = num % 10
                digit_sum = digit_sum + temp
                temp= num // 10
                num  = temp
        return abs(ele_sum - digit_sum)
                
obj = Solution()
result = obj.differenceOfSum([1,15,6,3])
print(result)
```

```python
#14. Longest Common Prefix
from typing import List
class Solution:
    def longestCommonPrefix(self, strs: List[str]) -> str:
        if not strs:
            return ""
        prefix = strs[0]
        for s in strs[1:]:
            while not s.startswith(prefix):
                prefix = prefix[:-1]
                if not prefix:
                    return ""
        return prefix
obj = Solution()
result = obj.longestCommonPrefix(["flower","flow","flight"])
print(result)
```

```python
#2461. Maximum Sum of Distinct Subarrays With Length K
from collections import defaultdict

class Solution:
    def maximumSubarraySum(self, nums, k):
        freq = defaultdict(int)
        left = 0
        window_sum = 0
        max_sum = 0
        for right in range(len(nums)):
            # add new number
            freq[nums[right]] += 1
            window_sum += nums[right]
            

            # if size > k → remove left
            if right - left + 1 > k:
                freq[nums[left]] -= 1
                window_sum -= nums[left]

                if freq[nums[left]] == 0:
                    del freq[nums[left]]

                left += 1

            # if size == k and all unique
            if right - left + 1 == k and len(freq) == k:
                max_sum = max(max_sum, window_sum)

        return max_sum
obj = Solution()
result = obj.maximumSubarraySum([1,5,4,2,9,9,9],3)
print(result)
```
