```python
# Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.
# You may assume that each input would have exactly one solution, and you may not use the same element twice.
# You can return the answer in any order.

from typing import List

class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        data = {}
        for i, num in enumerate(nums):
            newNum = target - num
            print(newNum)
            if newNum in data:
                print("for understand",data[newNum], i)
                return [data[newNum], i]
                
            
            data[num] = i 
            print("data",data, i)

nums = [7, 15, 2 ,11]
target = 9
obj = Solution()
result = obj.twoSum(nums, target)
print("result:- ",result)


```

```python
#3. Longest Substring Without Repeating Characters
class Solution:
    def lengthOftheLongestSubstring(self, s:str)-> int:
        char_set = set()
        left = 0
        max_len = 0
        for right in range(len(s)):
            while s[right] in char_set:
                print(left)
                char_set.remove(s[left])
                left += 1
            char_set.add(s[right])
            print(char_set)
            max_len = max(max_len, right - left +1)
        return max_len
        
       
        

obj = Solution()
result = obj.lengthOftheLongestSubstring("pwwkew")
print("total:",result)
```

```python
#189. Rotate Array
from typing import List
class Solution:
    def rotate(self, nums: List[int], k: int) -> None:
      
        n = len(nums)
        k = k%n
        print(k)
        temp = nums[n-k:]
        print(temp)        
        for i in range(n-k-1,-1,-1):
            nums[i+k]  = nums[i]
        for i in range(k):
            nums[i] =temp[i]

        return nums
        
        

obj = Solution()
result = obj.rotate([1,2,3,4,5,6,7],3)
print(result)
```

```python
#189. Rotate Array

from typing import List
class Solution:
    def rotate(self, nums: List[int], k: int) -> None:
        n = len(nums)
        k = k % n
        print(k, nums)
        nums.reverse() 
        print(nums)
        nums[:k] = reversed(nums[:k])
        print(nums[:k])
        nums[k:] = reversed(nums[k:])
        print(nums[k:])
        return nums
        
obj = Solution()
result = obj.rotate([1,2,3,4,5,6,7],3)
print(result)
```

```python
#189. Rotate Array

from typing import List
class Solution:
    def rotate(self, nums: List[int], k: int) -> None:
        n = len(nums)
        k = k % n
        roteted =[0]*n
        print(k, nums)
        for i in range(n):
            shift= (i + k)%n
            roteted[shift] = nums[i]
            
        for i in range(n):
            nums[i] = roteted[i]
        print(nums)
      
obj = Solution()
result = obj.rotate([1,2,3,4,5,6,7],2)
```

```python
#121. Best Time to Buy and Sell Stock
from typing import List
class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        maxprofit = 0
        minprice = 999999
        for i in range(len(prices)):
            #buying
            minprice = min(minprice,prices[i])
            #selling
            maxprofit = max(maxprofit,prices[i] - minprice)
        return maxprofit
        
obj = Solution()
result = obj.maxProfit([7,1,5,3,6,4])
print(result)
```

```python
#121. Best Time to Buy and Sell Stock
from typing import List
class Solution:
     def maxProfit(self, prices: List[int]) -> int:
         n = len(prices)
         mx = 0
         ans = 0
         for i in range(n -1,-1 ,-1):
             mx = max(prices[i],mx)
             ans = max(ans,mx - prices[i])
         return ans
obj = Solution()
result = obj.maxProfit([7,1,5,3,6,4])
print(result)
```

```python
#905.Short Array By Parity
class Solution:
    def sortArrayParity(self, nums: List[int]):
        print(nums)
        even = []
        odd = []
        for i  in range(len(nums)):
            if nums[i] % 2 == 0:
                even.append(nums[i])
            else:
                odd.append(nums[i])
        print(even + odd)

obj = Solution()

obj.sortArrayParity([3,1,2,4])
```

```python
#2161 Partition Array According to Given Pivot
from typing import List
class Solution:
    def pivotArray(self,  nums:List[int],pivot: int):
        print(nums)
        big = []
        little = []
        equal = []
        for num in nums:
            if num > pivot :
                big.append(num)
            elif num < pivot:
                little.append(num)
            if num == pivot:
                 equal.append( num)
        print(little+equal +big)
                
obj = Solution()
obj.pivotArray([-3,4,3,2],2)
```

```python
#75. Short Color
from typing import List
class Solution:
    def sortColors(self, nums: List[int]) -> None:
       l = 0
       m = 0
       r = len(nums) -1
       print(nums)
       while m <= r:
           print(nums[m])
           if nums[m] == 0:
               nums[m],nums[l] == nums[l],nums[m]
               l+=1
               m+= 1
           elif nums[m] == 1:
               m += 1
           elif nums[m] == 2:
               nums[m],nums[r] == nums[r],nums[m]
               r -= 1
       print(nums)
       
obj = Solution()
obj.sortColors([2,0,2,1,1,0])
```

```python
#383
class Solution:
    def canConstruct(self, ransomNote: str, magazine: str) -> bool:
        freq = [0] * 26
        for ch in magazine:
            freq[ord(ch)- ord("a")] += 1
        for ch in ransomNote:
            index = ord(ch) - ord("a")
            if freq[index] == 0:
                return False
            freq[index] -= 1
        return True
obj = Solution()
result = obj.canConstruct("bb","bba")
print(result)
```
