```python
class BankAccount:
    def __init__(self, name , acc_num, bal):
        self.account_holder_name  = name
        self.account_number = acc_num
        self.balance = bal
  
    def withdraw(self,amount):
        bal = self.balance - amount
        print(f"Balance for {self.account_holder_name} Post withdrawing {amount} amount , update balace is {bal}")
        self.balance = bal
    def deposit(self,amount):
        bal = self.balance + amount
        print(f"Balance for {self.account_holder_name} Post deposit {amount} amount , update balace is {bal}")
        self.balance = bal
    def  display(self):
        print("-------------------------")
        print("Acc name | number | balance")
        print(f"{self.account_holder_name} || {self.account_number} || {self.balance}")
```

```python
vishal = BankAccount("vishal","13784394",3498982)
ajeet = BankAccount("ajeet", "243245",3242345)

vishal.withdraw(3498982)
ajeet.withdraw(244)
vishal.deposit(234)
vishal.withdraw(100)
```

```python
# Use a constructor to initialize:
# length
# width
# Create methods:
# area()
# perimeter()
# Create multiple rectangle objects with different dimensions and print their area and perimeter.
class Rectangle:
    def __init__(self,length, widht):
        self.length = length
        self.widht = widht
    def area(self):
        area1 =  self.length * self.widht
        print(f"Area of reactangle is {area1}")
        
    def perimeter(self):
        perimeters  = 2 * (self.length + self.widht)
        print(F"Area of parameter is {perimeters}")
        
    
ract1 = Rectangle(10,5)
ract2 = Rectangle(14,6)
ract1.area()
ract2.area()
ract2.perimeter()
```

```python
# Constructor should initialize:
# employee_id
# name
# basic_salary
# Create a method calculate_salary() that adds:
# 20% HRA
# 10% DA
# Display total salary for the employee.
class Employee:
    def __init__(self,employee_id, name, basic_salary):
        self.employee_id = employee_id
        self.name = name
        self.basic_salary = basic_salary
    def calculate_salary(self):
        hra = 0.20 * self.basic_salary
        da = 0.10 * self.basic_salary
        total_salary = self.basic_salary + hra + da
        print(f"Employee ID: {self.employee_id}")
        print(f"Name: {self.name}")
        print(f"Total Salary: {total_salary}")

emp1 = Employee(1,"vishal",2000000)
emp1.calculate_salary()
        
```

```python
# Use a constructor to initialize:
# title
# author
# price
# Create a method apply_discount(percent) that reduces the book price.
# Create a method display_book() to show book details.
# Create at least two book objects and apply different discounts.
class Book:
    def __init__(self, title,author,price):
        self.title = title
        self.author = author
        self.price = price
    def apply_discount(self, percent):
        dicount_amount = ( percent /100) * self.price
        self.price -=  dicount_amount

    def display_book(self):
        print(f"Title: {self.title}")
        print(f"Author: {self.author}")
        print(f"Price: {self.price}")
        print("-" * 20)
book1 = Book("Clean Code", "Robert C. Martin", 500)
book2 = Book("Python Crash Course", "Eric Matthes", 600)
book1.apply_discount(10)  
book2.apply_discount(20)  
book1.display_book()
book2.display_book()
```
