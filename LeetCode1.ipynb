```python
#121. Best Time to Buy and Sell Stock
from typing import List
class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        maxprofit = 0
        minprice = 999999
        for i in range(len(prices)):
            #buying
            minprice = min(minprice,prices[i])
            #selling
            maxprofit = max(maxprofit,prices[i] - minprice)
        return maxprofit
        
obj = Solution()
result = obj.maxProfit([7,1,5,3,6,4])
print(result)
```

```python
#121. Best Time to Buy and Sell Stock
from typing import List
class Solution:
     def maxProfit(self, prices: List[int]) -> int:
         n = len(prices)
         mx = 0
         ans = 0
         for i in range(n -1,-1 ,-1):
             mx = max(prices[i],mx)
             ans = max(ans,mx - prices[i])
         return ans
obj = Solution()
result = obj.maxProfit([7,1,5,3,6,4])
print(result)
```

```python
#383.Ransom Note
class Solution:
    def canConstruct(self, ransomNote: str, magazine: str) -> bool:
        freq = [0] * 26
        for ch in magazine:
            freq[ord(ch)- ord("a")] += 1
        for ch in ransomNote:
            index = ord(ch) - ord("a")
            if freq[index] == 0:
                return False
            freq[index] -= 1
        return True
obj = Solution()
result = obj.canConstruct("bb","bba")
print(result)
```

```python
#383. Ransom Note 
class Solution:
    def canConstruct(self, ransomNote: str, magazine: str) -> bool:
        mz = dict()
        for ch in magazine:
            if ch in mz:
                mz[ch] += 1
            else:
                mz[ch] = 1
        
        for ch in ransomNote:
            if ch not in mz or mz[ch] == 0:
                return False
            mz[ch] -= 1    
        return True    

obj = Solution()
result = obj.canConstruct("aa","ab")
print(result)
```

```python
# 49 Group Anagrams
from collections import defaultdict
class Solution:
    def groupAnagrams(self,strs:list[str]):
        mp = defaultdict(list)
        for word in strs:
            key = "".join(sorted(word))
            mp[key].append(word)
        return list(mp.values())
            
obj = Solution()
result = obj.groupAnagrams(["eat","tea","tan","ate","nat","bat"])
print(result)
```

```python
# 49 Group Anagrams
from collections import defaultdict
class Solution:
    def groupAnagrams(self,strs:list[str]):
        mp = defaultdict(list)
        for s in strs:
            count = [0] * 26
            for c in s:
                count[ord(c) - ord('a')] += 1
            mp[tuple(count)].append(s)
        return list(mp.values())
            
obj = Solution()
result = obj.groupAnagrams(["eat","tea","tan","ate","nat","bat"])
print(result)
```

```python
#88 merge shorted array
class Solution:
    def merge(self,nums1:list[int],m:int,nums2:list[int],n):
        i = m - 1       # pointer for nums1
        j = n - 1       # pointer for nums2
        k = m + n - 1   # pointer for merged array

        while i >= 0 and j >= 0:
            if nums1[i] > nums2[j]:
                nums1[k] = nums1[i]
                i -= 1
            else:
                nums1[k] = nums2[j]
                j -= 1
            k -= 1
        # # If nums2 still has elements left
        # while j >= 0:
        #     nums1[k] = nums2[j]
        #     j -= 1
        return nums1
            
obj = Solution()
result = obj.merge([1,2,3,7,0,0],3,[2,5,6],3)
print(result)
```

```python
#26 Remove Duplicates from Sorted Array
class Solution:
    def removeDuplicates(self,nums:list[int]):
        k = 0
        for i in range(1,len(nums)):
            if nums[k] != nums[i]:
                nums[k+1] = nums[i]
                k+=1
        return k +1
            
obj = Solution()
result = obj.removeDuplicates([1,1,2])
print(result)
```

```python
#169 Majority Element
class Solution:
    def majorityElement(self,nums:list[int]):
        cnt = 0
        for i in range(len(nums)):
            if cnt == 0:
                res = nums[i]
            if nums[i] == res:
                cnt += 1
            else:
                cnt -= 1
        return res
obj = Solution()
result = obj.majorityElement([2,3,3,2])
print(result)
```

```python
#11. Container With Most Water
# Formula Water = min(left_height, right_height) Ã— distance_between_lines
class Solution:
    def maxArea(self,height:list[int]):
        left = 0
        right = len(height) - 1
        max_water = 0
        while left < right:
            width = right - left
            h = min(height[left] , height[right])
            area = width * h
            max_water = max(max_water, area)
            if height[left] < height[right]:
                left += 1
            else:
                right -= 1
        return max_water
        
            
obj = Solution()
result = obj.maxArea([1,8,6,2,5,4,8,3,7])
print(result)
```
